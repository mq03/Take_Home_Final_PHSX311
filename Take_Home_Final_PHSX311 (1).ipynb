```python
#Fourier.py
#PHSX311 HW20 code
#Code to calculate fourier series and fourier transforms
#import necessary packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fftfreq, fft
import os
```

```python
#read in data
dat = np.load('FinalData.npz')
t = dat['t'] #time
y = dat['y'] #y-position
```

```python
#plot data
plt.clf()
plt.plot(t, y, color = 'black', marker = '.', linestyle = 'None')
plt.xlabel('Time(s)')
plt.ylabel('Position(m)')
plt.show()
```

```python
#values to be used for frequency array set up

N = np.size(y) #number of points in dataset
dt = t[1]-t[0] #time-spacing between datapoints
```

```python
#Set up arrays for periodogram

freq = fftfreq(N, dt)[:N//2] #frequencies
pow = 2.0/N * np.abs(fft(y)[:N//2]) #power
```

```python
#cut out indices of frequencies that can't be evaluated

maxT = np.max(t)-np.min(t) #max period that can be evaluated in dataset
minf = 1/maxT #min frequency that can be evaluated in dataset
out = np.where(freq < minf)[0] #indices to be cut out of data set
freq = np.delete(freq, out) #cut out time array here
pow = np.delete(pow, out) #cut out of power array here
```

```python
#Plot the periodogram

plt.clf()
plt.plot(freq, pow)
plt.xlabel('Frequency(Hz)')
plt.ylabel('Power(W)')
plt.title('Peak frequencies = 50, 120, 400')
plt.grid()
plt.show()
```
